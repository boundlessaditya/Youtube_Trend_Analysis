```python
import pandas as pd
import numpy as np
```

```python
def fill_null(df):
    for i in range(len(df)):
        str = df.columns[i]
        if df.iloc[:,i].dtypes != df.iloc[:,0].dtypes:
            df[str].fillna(df[str].mean(), inplace = True)
        else:
            df[str].fillna(df[str].value_counts().index[0], inplace = True)
```

```python
Data = pd.read_csv("youtube_stats.csv",index_col = 0,encoding_errors = 'replace')
df = Data.copy()
```
